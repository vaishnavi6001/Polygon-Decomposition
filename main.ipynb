```python
import matplotlib.pyplot as plt

with open("out1.txt", "r") as f:
    lines = f.readlines()

coordinates = []
 
for line in lines:
    x1, y1, x2, y2 = map(int, line.strip().split())
    coordinates.append(((x1, y1), (x2, y2)))

fig, ax = plt.subplots()
color = 'k'

for coord in coordinates:
    ax.plot([coord[0][0], coord[1][0]], [coord[0][1], coord[1][1]], color=color)
    ax.text(coord[0][0]+0.1, coord[0][1]+0.1, '({}, {})'.format(coord[0][0], coord[0][1]), color='b')
    ax.plot(coord[0][0], coord[0][1], 'ro')
    
plot_margin = 1

x0, x1, y0, y1 = plt.axis()
plt.axis((x0 - plot_margin,
          x1 + plot_margin,
          y0 - plot_margin,
          y1 + plot_margin))
plt.show()
```
